```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph to represent neural circuitry
G = nx.DiGraph()

# Add nodes representing brain regions
regions = {
    "Hippocampus": (-4, 0),
    "Entorhinal Cortex": (-2, 2),
    "Retrosplenial Cortex": (0, 0),
    "Prefrontal Cortex": (2, -2),
    "Amygdala": (4, 0)
}
G.add_nodes_from(regions.keys())

# Add edges representing connectivity (arrows for direction)
edges = [
    ("Hippocampus", "Entorhinal Cortex"),
    ("Entorhinal Cortex", "Retrosplenial Cortex"),
    ("Hippocampus", "Prefrontal Cortex"),
    ("Prefrontal Cortex", "Amygdala"),
    ("Retrosplenial Cortex", "Amygdala")
]
G.add_edges_from(edges)

# Simulate Alzheimer's by marking specific connections as damaged
damaged_edges = [("Hippocampus", "Entorhinal Cortex"), ("Entorhinal Cortex", "Retrosplenial Cortex")]

# Visualize the neural circuitry
plt.figure(figsize=(12, 8))
pos = regions

# Draw intact connections (blue)
intact_edges = [edge for edge in edges if edge not in damaged_edges]
nx.draw_networkx_edges(G, pos, edgelist=intact_edges, edge_color="blue", arrowstyle="->", arrowsize=15, width=2)

# Draw damaged connections (red)
nx.draw_networkx_edges(G, pos, edgelist=damaged_edges, edge_color="red", arrowstyle="->", arrowsize=15, width=2)

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color="skyblue", edgecolors="black", linewidths=1.5)

# Add labels to nodes
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold", font_color="black")

# Add legend
plt.text(-5, -3, "Intact Connections (Blue)", fontsize=10, color="blue", ha="left")
plt.text(-5, -4, "Damaged Connections (Red)", fontsize=10, color="red", ha="left")

# Add title
plt.title("Neural Circuitry in Alzheimer's Disease", fontsize=16, fontweight="bold")

# Show the visualization
plt.axis("off")
plt.show()
```
